```python
import pandas as pd
```

```python
#데이터 불러오기

path = r"C:\Users\82106\Desktop/리뷰.csv"
df = pd.read_csv(path)
df
```

```python
#필요한 열만 추출

df = df.loc[:, ["reviewer_id","reviewer_name","comments_KO"]]
df
```

```python
df.info()
```

```python
#개행문자 제거
# \n(개행문자) -> ""(공백)

df['comments_KO'] = df['comments_KO'].apply(lambda x: x.replace('\n',''))

#HTML코드 제거
# <br/> -> ""(공백)
df['comments_KO'] = df['comments_KO'].apply(lambda x: x.replace('<br/>',''))
```

```python
#특수문자 제거

import re
def cleanText(readData):
 
    
    text = re.sub('[-=+,#/\?:;^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》❤]', '', readData)
    return text

df['comments_KO'] = df['comments_KO'].apply(cleanText)
```

```python
#개별 국문 제거

def cleanText2(readData):
 
    text = re.sub('[ㄱㄴㄷㄹㅁㅂㅅㅇㅈㅊㅋㅌㅍㅎㅃㅉㄸㄲㅆㅛㅕㅑㅐㅔㅗㅓㅏㅣㅜㅠㅡ]', '', readData)
    return text
df['comments_KO'] = df['comments_KO'].apply(cleanText2)
```

```python
#전처리 알파벳 제거(프랑스어 등 언어 제거) -> 이름, 공항이름 등 모두 지워짐
#한자 및 깨진 문자는 어떻게? (ex 뚫 섺 등)
```

```python
#알파벳 제거 (불어, 기타 외국어 제거)

def cleanText3(readData):
 
    text = re.sub('[abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ]', '', readData)
    return text
df['comments_KO'] = df['comments_KO'].apply(cleanText3)
```
